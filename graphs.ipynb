# Graphs for the paper

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
```

# SSIM Vs. SNR 

```python
SNR = [0,5,10,15,20]
# SSIM values and standard deviations
MoDL = np.array([0.8267, 0.8767, 0.9062, 0.9329, 0.9446])
ViT = np.array([0.9264, 0.9446, 0.9583, 0.9680, 0.9730])
ViT_Fuser = np.array([0.9476, 0.9555, 0.9648, 0.9716, 0.9760])
ViT_Fuser_Hybrid = np.array([0.9487, 0.9517, 0.9608, 0.9683, 0.9726])
ViT_Hybrid = np.array([0.9167, 0.9371, 0.9521, 0.9635, 0.9695])

MoDL_std = np.array([0.0112, 0.0089, 0.0072, 0.0061, 0.0065])
ViT_std = np.array([0.0077, 0.0049, 0.0041, 0.0041, 0.0041])
ViT_Fuser_std = np.array([0.0054, 0.0038, 0.0035, 0.0032, 0.0033])
ViT_Fuser_Hybrid_std = np.array([0.0045, 0.0039, 0.0035, 0.0036, 0.0039])
ViT_Hybrid_std = np.array([0.0079, 0.0057, 0.0049, 0.0043, 0.0048])

# Colors for each plot
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
#ax.set_facecolor('lightgray')
ax.set_facecolor((1, 1, 0.8, 0.3))

# Plot MoDL with consistent color for line and std
ax.plot(SNR, MoDL, '-^', color=colors[0], label='MoDL - MSE loss', linewidth=1.5)
ax.fill_between(SNR, MoDL - MoDL_std, MoDL + MoDL_std, color=colors[0], alpha=0.2)
"""
# Plot ViT with consistent color for line and std
ax.plot(SNR, ViT, '-o', color=colors[1], label='ViT - MSE loss', linewidth=1.5)
ax.fill_between(SNR, ViT - ViT_std, ViT + ViT_std, color=colors[1], alpha=0.2)

# Plot ViT-Fuser with consistent color for line and std
ax.plot(SNR, ViT_Fuser, '-s', color=colors[2], label='ViT-Fuser - MSE loss', linewidth=1.5)
ax.fill_between(SNR, ViT_Fuser - ViT_Fuser_std, ViT_Fuser + ViT_Fuser_std, color=colors[2], alpha=0.2)
"""
# Plot ViT-Hybrid with consistent color for line and std
ax.plot(SNR, ViT_Hybrid, '-d', color=colors[3], label='ViT - Hybrid loss', linewidth=1.5)
ax.fill_between(SNR, ViT_Hybrid - ViT_Hybrid_std, ViT_Hybrid + ViT_Hybrid_std, color=colors[3], alpha=0.2)

# Plot ViT-Fuser-Hybrid with consistent color for line and std
ax.plot(SNR, ViT_Fuser_Hybrid, '-p', color=colors[4], label='ViT-Fuser - Hybrid loss', linewidth=1.5)
ax.fill_between(SNR, ViT_Fuser_Hybrid - ViT_Fuser_Hybrid_std, ViT_Fuser_Hybrid + ViT_Fuser_Hybrid_std, 
                 color=colors[4], alpha=0.2)


# Customize the plot
ax.set_xlabel('SNR [dB]', fontsize=12)
ax.set_ylabel('SSIM', fontsize=12)
ax.set_title('SSIM Metric as a Function of SNR', fontsize=14)
ax.legend(loc='best', fontsize=10)
ax.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

# CMMD Vs. SNR

```python
# Data
SNR = np.array([0, 3, 5, 10, 15, 20])  # SNR values (x-axis)

# CMMD metric values for each method
MoDL_MSE = np.array([1.014, 0.595, 0.362, 0.213, 0.183, 0.146])

MoDL_Hybrid = np.array([1.214,0.838,0.680,0.382,0.284,0.231])

ViT_Hybrid = np.array([0.168, 0.142, 0.130, 0.075, 0.054, 0.033])

ViT_Fuser_Hybrid = np.array([0.051, 0.065, 0.055, 0.044, 0.038, 0.022])


# Colors for each plot
colors = ['tab:blue', 'tab:orange', 'tab:green']

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot MoDL - MSE Loss 
ax.plot(SNR, MoDL_MSE, '-o', color=colors[0], label='MoDL - MSE loss', linewidth=1.5)

# Plot MoDL - Hybriid Loss 
ax.plot(SNR, MoDL_Hybrid, '--o', color=colors[0], label='MoDL - Hybrid loss', linewidth=1.5)

# Plot ViT - Hybrid Loss
ax.plot(SNR, ViT_Hybrid, '-d', color=colors[1], label='ViT - Hybrid loss', linewidth=1.5, markersize=6)

# Plot ViT-Fuser - Hybrid Loss
ax.plot(SNR, ViT_Fuser_Hybrid, '-^', color=colors[2], label='ViT-Fuser - Hybrid loss', linewidth=1.5, markersize=6)

# Customize the plot
ax.set_xlabel('SNR [dB]', fontsize=12)
ax.set_ylabel('CMMD Metric Value', fontsize=12)
ax.set_title('CMMD Metric as a Function of SNR', fontsize=14)
ax.legend(loc='upper right', fontsize=10)
ax.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

# FID Vs. SNR

```python

# Data
SNR = np.array([0, 3, 5, 10, 15, 20])  # SNR values (x-axis)

# FID metric values for each method
MoDL_MSE = np.array([0.7178, 0.1968, 0.0807, 0.0253, 0.0262, 0.0135])
MoDL_Hybrid = np.array([0.942, 0.5102, 0.4286, 0.0958, 0.0401, 0.0162])
ViT_Hybrid = np.array([0.1490, 0.1236, 0.1094, 0.0737, 0.0334, 0.0208])
ViT_Fuser_Hybrid = np.array([0.0459, 0.0500, 0.0389, 0.0322, 0.0229, 0.0114])

# Colors for each method
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot MoDL - MSE Loss
ax.plot(SNR, MoDL_MSE, '-o', color=colors[0], label='MoDL - MSE loss', linewidth=1.5, markersize=6)

# Plot MoDL - Hybrid Loss
ax.plot(SNR, MoDL_Hybrid, '--o', color=colors[0], label='MoDL - Hybrid loss', linewidth=1.5, markersize=6)

# Plot ViT - Hybrid Loss
ax.plot(SNR, ViT_Hybrid, '-d', color=colors[1], label='ViT - Hybrid loss', linewidth=1.5, markersize=6)

# Plot ViT-Fuser - Hybrid Loss
ax.plot(SNR, ViT_Fuser_Hybrid, '-^', color=colors[2], label='ViT-Fuser - Hybrid loss', linewidth=1.5, markersize=6)

# Customize the plot
ax.set_xlabel('SNR [dB]', fontsize=12)
ax.set_ylabel('FID Metric Value', fontsize=12)
ax.set_title('FID Metric as a Function of SNR', fontsize=14)
ax.legend(loc='upper right', fontsize=10)
ax.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
